```python
import subprocess
import sys
from pathlib import Path


ROOT = Path.cwd().resolve().parent


STEPS = [
    [sys.executable, "-m", "jupyter", "execute", "codigo/2_feature_engineering.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/3_catboost_best_submit.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/5_catboost_feature_selection.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/7_catboost_top60_hypercv.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/11_catboost_top60_postprocess.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/13_extratrees_models.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/14_extratrees_blends.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/16_hardcode_genre.ipynb"],
    [sys.executable, "-m", "jupyter", "execute", "codigo/17_second_subgenre_probe.ipynb"],
]
```

```python
def run_step(command):
    print("\n" + "=" * 100)
    print("Ejecutando:", " ".join(command))
    print("=" * 100)
    subprocess.run(command, cwd=ROOT, check=True)
```

```python
def main():
    train_path = ROOT / "datos_raw" / "base_train.csv"
    test_path = ROOT / "datos_raw" / "base_val.csv"
    if not train_path.exists() or not test_path.exists():
        raise FileNotFoundError(
            "Faltan datos_raw/base_train.csv y/o datos_raw/base_val.csv."
        )

    for command in STEPS:
        run_step(command)

    final_path = ROOT / "Submits" / "hardcode_rank_neo60_progressive_down_24.csv"
    print("\nModelo final reproducible generado:")
    print(final_path)
```

```python
if __name__ == "__main__":
    main()
```
